# Question 6

Pesquise a apresente uma ou mais aplicações da rede SOM.

Authors: 
  - [Morsinaldo de Azevedo Medeiros](https://github.com/Morsinaldo)
  - [Thaís de Araújo de Medeiros](https://github.com/thaisaraujo2000)

Article Built Environment, Psychosocial Factors and Active Commuting to School in Adolescents: Clustering a Self-Organizing Map Analysis

Source: https://www.mdpi.com/1660-4601/16/1/83

O artigo analisa a conexão entre o ambiente construído, fatores psicossociais e o deslocamento ativo para a escola (ACS) em adolescentes. A pesquisa, realizada em Valência, Espanha, com 465 adolescentes, utilizou uma análise de mapa auto-organizado para identificar padrões de comportamento e relações entre esses elementos. Foram medidos variáveis como ACS, barreiras ao ACS, autoeficácia física, apoio social e dados sociodemográficos.

Os dados foram coletados por meio de questionários. Além disso, a distância da rede viária até a escola, a densidade residencial e a densidade de interseções de ruas foram calculadas usando o Sistema de Informações Geográficas. O método k-means foi utilizado para agrupar os neurônios em clusters com base nas características das variáveis de entrada, e o teste de Kruskal-Wallis foi aplicado para comparar as variáveis entre esses clusters.

Os resultados mostram que níveis mais baixos de ACS estão associados a ambientes construídos menos favoráveis, como baixa densidade residencial e conectividade viária, maiores distâncias até a escola, variáveis psicossociais desfavoráveis, como baixa autoeficácia física e apoio social médio para ACS, e fácil acesso a transporte motorizado privado.
